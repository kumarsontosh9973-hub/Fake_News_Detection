```python
def wordopt(text):
    text=text.lower()
    text=re.sub('\[.*?\]' , '' ,text)
    text = re.sub('\\W', ' ', text)

    text=re.sub('https?://\s+|www\.\s+' , '' , text)
    text=re.sub('<.*?>+' , '' , text)
    text=re.sub('[%s]' %re.escape(string.punctuation) , '' , text)
    text=re.sub('\n' , '' , text)
    text=re.sub('\w*\d\w*' , '' , text)
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x  , y , test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization=TfidfVectorizer()
```

```python
xv_train=vectorization.fit_transform(x_train)
xv_test=vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression()
```

```python
LR.fit(xv_train , y_train)
```

```python
predicted_LR=LR.predict(xv_test)
LR.score(xv_test , y_test)
```

```python
print(classification_report(y_test , predicted_LR))
```

```python
def output_lable(n):
    if n==0:
        return "fake news"
    elif n==1:
        return "not a fake news"
    
```

```python
def manual_testing(news):
    testing_news = {"text": [news]}
    
    news_def_testing = pd.DataFrame(testing_news)
    
    news_def_testing['text'] = news_def_testing['text'].apply(wordopt)
    
    news_x_testing = news_def_testing['text']
    
    news_xv_testing = vectorization.transform(news_x_testing)
    
    predicted_LR = LR.predict(news_xv_testing)
    predicted_DT = DT.predict(news_xv_testing)
    predicted_GBSC = GBSC.predict(news_xv_testing)
    predicted_RFC = RFC.predict(news_xv_testing)
    
    return print(
        "\n\nLR prediction:{}\nDT prediction:{}\nGBSC prediction:{}\nRFC prediction:{}".format(
            output_lable(predicted_LR[0]),
            output_lable(predicted_DT[0]),
            output_lable(predicted_GBSC[0]),
            output_lable(predicted_RFC[0])
        )
    )
```
